```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"


# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
# mouse_data 
clinical_data = pd.read_csv(clinical_trial_data_to_load)
# clinical_data
# # Combine the data into a single dataset
data_complete = pd.merge(clinical_data, mouse_data, how="left", on=["Mouse ID", "Mouse ID"]) #on=["school_name", "school_name"])

# # Display the data table for preview

data_complete
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

mean_tumor_volume_drug=data_complete.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

mean_tumor_volume_drug
# mean_tumor_volume_timepoint=data_complete.groupby(["Timepoint"]).mean()["Tumor Volume (mm3)"]
# mean_tumor_volume_timepoint

# Convert to DataFrame
merge_data=pd.DataFrame(mean_tumor_volume_drug.reset_index())
merge_data
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
mean_tumor_volume_SEM=data_complete.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

mean_tumor_volume_SEM
# Convert to DataFrame
merge_data_SEM=pd.DataFrame(mean_tumor_volume_SEM.reset_index())
# Preview DataFrame

merge_data_SEM
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatted_volume = merge_data.pivot(index = 'Timepoint', columns='Drug', values='Tumor Volume (mm3)')
reformatted_SEM = merge_data_SEM.pivot(index = 'Timepoint', columns='Drug', values='Tumor Volume (mm3)')
# reformatted_SEM
reformatted_volume
# Preview that Reformatting worked
```

```python
# Generate the Plot (with Error Bars)
x_axis = np.arange(0, 50, 5)
# reformatted_volume.plot(kind="line", x="Timepoint", y="Capomuline", grid=True, figsize=(20,10),
#               title="Tumor sizes Vs. Drugs")
plt.errorbar (reformatted_volume.index,reformatted_volume["Capomulin"],yerr=reformatted_SEM["Capomulin"], color="red",marker='o', markersize=7, linewidth=.5, linestyle='dotted')
plt.errorbar (reformatted_volume.index,reformatted_volume["Infubinol"],yerr=reformatted_SEM["Infubinol"], color="b",marker='^', markersize=7, linewidth=.5, linestyle='dotted')
plt.errorbar (reformatted_volume.index,reformatted_volume["Ketapril"],yerr=reformatted_SEM["Ketapril"], color="g",marker='s', markersize=7, linewidth=.5, linestyle='dotted')
plt.errorbar (reformatted_volume.index,reformatted_volume["Placebo"],yerr=reformatted_SEM["Placebo"], color="black",marker='d', markersize=7, linewidth=.5, linestyle='dotted')
plt.title("Tumor sizes Vs. Drugs")
plt.grid(True)
plt.xlabel("Treatment duration (days)")
plt.ylabel("Tumor Volume mm3")
plt.legend(loc='upper left')

# Save the Figure

plt.savefig("../Images/Tumor_volume_SEM.png")

plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_Metsite=data_complete.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
mean_Metsite
# Convert to DataFrame
merge_data_Metsite=pd.DataFrame(mean_Metsite.reset_index())
# Preview DataFrame
merge_data_Metsite
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

mean_SEM=data_complete.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]
mean_SEM
# Convert to DataFrame
metsize_SEM=pd.DataFrame(mean_SEM.reset_index())
# Preview DataFrame
metsize_SEM
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatted_metsize = merge_data_Metsite.pivot(index = 'Timepoint', columns='Drug', values='Metastatic Sites')
reformatted_metsize_SEM = metsize_SEM.pivot(index = 'Timepoint', columns='Drug', values='Metastatic Sites')
reformatted_metsize_SEM
# Preview that Reformatting worked
# reformatted_metsite
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar (reformatted_metsize.index,reformatted_metsize["Capomulin"],yerr=reformatted_metsize_SEM["Capomulin"], color="red",marker='o', markersize=7, linewidth=.5, linestyle='dotted')
plt.errorbar (reformatted_metsize.index,reformatted_metsize["Infubinol"],yerr=reformatted_metsize_SEM["Infubinol"], color="b",marker='^', markersize=7, linewidth=.5, linestyle='dotted')
plt.errorbar (reformatted_metsize.index,reformatted_metsize["Ketapril"],yerr=reformatted_metsize_SEM["Ketapril"], color="g",marker='s', markersize=7,  linewidth=.5, linestyle='dotted')
plt.errorbar (reformatted_metsize.index,reformatted_metsize["Placebo"],yerr=reformatted_metsize_SEM["Placebo"], color="black",marker='d', markersize=7,  linewidth=.5, linestyle='dotted')
plt.title("Met. sites Vs. Drugs")
plt.grid(True)
plt.xlabel("Treatment Duration (days)")
plt.ylabel("Met. sites")
plt.legend(loc='upper left')



# Save the Figure
plt.savefig("../Images/Metsite_SEM.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count=data_complete.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]
mouse_count
# Convert to DataFrame
merge_mouse_count=pd.DataFrame(mouse_count.reset_index())
# Preview DataFrame
merge_mouse_count
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatted_survival = merge_mouse_count.pivot(index = 'Timepoint', columns='Drug', values='Mouse ID')
# Preview the Data Frame
reformatted_survival
```

```python
# survival plot
plt.plot(reformatted_survival.index, (reformatted_survival["Capomulin"]/25)*100, color="red",marker='o', markersize=7, linewidth=.5, linestyle='dotted')
plt.plot(reformatted_survival.index, (reformatted_survival["Infubinol"]/25)*100,color="b",marker='^', markersize=7, linewidth=.5, linestyle='dotted')
plt.plot(reformatted_survival.index, (reformatted_survival["Ketapril"]/25)*100, color="g",marker='s', markersize=7,  linewidth=.5, linestyle='dotted')
plt.plot(reformatted_survival.index, (reformatted_survival["Placebo"]/25)*100, color="g",marker='s', markersize=7,  linewidth=.5, linestyle='dotted')


plt.show
```
